```python
import pandas as pd
import numpy as np
data=pd.read_csv("crime_data.csv")
```

```python
data.head(5)
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
scaler =StandardScaler()
scaled_features = scaler.fit_transform(data)
```

```python
model=KMeans(n_clusters=3)
model.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
model=AgglomerativeClustering(n_clusters=3,affinity="l2",linkage="complete")

model.fit_predict(scaled_features)
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
model=AgglomerativeClustering(n_clusters=3,affinity="l2",linkage="average")

model.fit_predict(scaled_features)
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
model=AgglomerativeClustering(n_clusters=3,affinity="l2",linkage="single")

model.fit_predict(scaled_features)
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
model=DBSCAN(min_samples=2,eps=0.1)

model.fit_predict(scaled_features)
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
model=DBSCAN(min_samples=3,eps=0.1)

model.fit_predict(scaled_features)
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(data['Murder'],data['Assault'],c=model.fit_predict(scaled_features),cmap='viridis')
```

```python
nc=range(1,11)
l=[]
for i in nc:

    model=KMeans(n_clusters=i)
    model.fit(scaled_features)
    l.append(model.inertia_)
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(nc,l,s=100,color="black")
plt.plot(nc,l,linestyle="-.")
plt.xticks(nc)
plt.show()
```

Silhoutee

```python
from sklearn.metrics import silhouette_score
```

```python
nc=range(2,11)
l=[]
for i in nc:

    model=AgglomerativeClustering(n_clusters=i,affinity='l2',linkage="complete")
    model.fit_predict(scaled_features)
    l.append(silhouette_score(scaled_features,model.fit_predict(scaled_features)))
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(nc,l,s=100,color="black")
plt.plot(nc,l,linestyle="-.")
plt.xticks(nc)
plt.show()
```
